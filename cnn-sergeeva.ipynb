```python
import os
os.environ['CUBLAS_WORKSPACE_CONFIG'] = ':4096:8'
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import librosa
import librosa.feature as F
import matplotlib.pyplot as plt
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from tqdm import tqdm
from enum import Enum

# Параметры
base_folder = "/kaggle/input/itmo-acoustic-event-detectin-2025"
train_folder = os.path.join(base_folder, "audio_train/train")
train_csv = os.path.join(base_folder, "train.csv")
train_pickle = "/kaggle/working/train.pickle"
labels_pickle = "/kaggle/working/labels.pickle"
random_state = 42
device = "cuda" if torch.cuda.is_available() else "cpu"
num_workers = os.cpu_count() if device == "cpu" else 0

# Установка seed для воспроизводимости
def set_random_state(random_state=42):
    torch.manual_seed(random_state)
    random.seed(random_state)
    np.random.seed(random_state)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(random_state)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_random_state(42)
# torch.use_deterministic_algorithms(True)

# Параметры обработки аудио
class SetType(Enum):
    TRAIN = 1
    TEST = 2

def extract_log_mel_feats(set_type, path_to_csv, path_to_files, out_path, labels_pickle, 
                          sr=22050, n_fft=2048, hop_length=512, n_mels=256, duration=3.0):
    feats = []
    if set_type == SetType.TRAIN:
        meta = pd.read_csv(path_to_csv, skiprows=1, names=['fname', 'label'])
        file_names = list(meta['fname'])
        labels = list(meta['label'])
        uniq_labels = np.sort(np.unique(labels))
        label_to_id = {label: i for i, label in enumerate(uniq_labels)}
        print('Total files:', len(file_names))
        for file_name, label in tqdm(zip(file_names, labels), total=len(file_names)):
            wav_data, sr = librosa.load(os.path.join(path_to_files, file_name), sr=sr, duration=duration)
            if len(wav_data) < sr * duration:
                wav_data = np.pad(wav_data, (0, int(sr*duration)-len(wav_data)))
            mel_spec = F.melspectrogram(y=wav_data, sr=sr, n_fft=n_fft, hop_length=hop_length, 
                                        n_mels=n_mels, fmax=sr//2)
            log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
            feats.append({'fname': file_name, 'feature': log_mel_spec, 'label_id': label_to_id[label]})
        pickle.dump(feats, open(out_path, 'wb'))
        pickle.dump(label_to_id, open(labels_pickle, 'wb'))
        return label_to_id
    else:
        for file_name in tqdm(os.listdir(path_to_files)):
            wav_data, sr = librosa.load(os.path.join(path_to_files, file_name), sr=sr, duration=duration)
            if len(wav_data) < sr * duration:
                wav_data = np.pad(wav_data, (0, int(sr*duration)-len(wav_data)))
            mel_spec = F.melspectrogram(y=wav_data, sr=sr, n_fft=n_fft, hop_length=hop_length,
                                        n_mels=n_mels, fmax=sr//2)
            log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
            feats.append({'fname': file_name, 'feature': log_mel_spec})
        pickle.dump(feats, open(out_path, 'wb'))

# Подготовка данных
extract_log_mel_feats(SetType.TRAIN, train_csv, train_folder, train_pickle, labels_pickle)
train_data = pickle.load(open(train_pickle, "rb"))
label_to_id = pickle.load(open(labels_pickle, "rb"))
id_to_label = [label for label, i in label_to_id.items()]

class EventDetectionDataset(Dataset):
    def __init__(self, features, labels=None, augment=False, device="cpu"):
        self.features = features
        self.labels = labels
        self.augment = augment
        self.device = device
        self.target_frames = 96  # Для 3-секундного аудио при 22050 Гц

    @staticmethod
    def prepare_shape(feature, target_frames=96):
        """Подготовка формы фичи с возможностью случайного смещения"""
        n_mels, n_frames = feature.shape  # shape: (n_mels, n_frames)
        
        # Добавляем измерение для канала
        feature = feature[np.newaxis, :, :]  # shape: (1, n_mels, n_frames)
    
        if n_frames < target_frames:
            pad_width = target_frames - n_frames
            padded = np.pad(feature, ((0, 0), (0, 0), (0, pad_width)), mode='constant')
            return padded
        elif n_frames > target_frames:
            start = np.random.randint(0, n_frames - target_frames)
            return feature[:, :, start:start+target_frames]
        else:
            return feature
            
    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        feat = self.features[idx]
        # Применяем prepare_shape к каждой фиче
        feat = self.prepare_shape(feat, self.target_frames)
        
        if self.augment:
            # Теперь feat трёхмерный
            _, _, n_frames = feat.shape
            if n_frames > self.target_frames:
                start = np.random.randint(0, n_frames - self.target_frames)
                feat = feat[:, :, start:start+self.target_frames]
        
        feat = feat[:, :, :self.target_frames]  # Обрезаем до нужного размера
        feat_tensor = torch.tensor(feat, dtype=torch.float32, device=self.device)
        
        if self.labels is not None:
            return feat_tensor, torch.tensor(self.labels[idx], device=self.device)
        return feat_tensor
        
# Разбиение данных
def stratified_split(data):
    features = [item['feature'] for item in data]
    labels = [item['label_id'] for item in data]
    return train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=random_state)

x_train, x_val, y_train, y_val = stratified_split(train_data)

# Создание датасетов и загрузчиков
train_dataset = EventDetectionDataset(x_train, y_train, augment=True, device=device)
val_dataset = EventDetectionDataset(x_val, y_val, device=device)
batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers)

# Улучшенная архитектура сети
class ImprovedNetwork(nn.Module):
    def __init__(self, num_classes=41):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=5, stride=2, padding=2),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 1),
            nn.Dropout(0.3),
            
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 1),
            nn.Dropout(0.3),
            
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 1),
            nn.Dropout(0.3),
            
            nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Dropout(0.5)
        )
        self.classifier = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

# Функция обучения
def train_model(model, train_loader, val_loader, criterion, optimizer, n_epochs=100):
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_acc = 0.0
    
    for epoch in range(n_epochs):
        # Тренировка
        model.train()
        running_loss = 0.0
        all_preds, all_labels = [], []
        
        for inputs, labels in tqdm(train_loader, desc=f'Epoch {epoch+1}/{n_epochs}'):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
        
        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = accuracy_score(all_labels, all_preds)
        history['train_loss'].append(epoch_loss)
        history['train_acc'].append(epoch_acc)
        
        # Валидация
        model.eval()
        running_loss = 0.0
        val_preds, val_labels = [], []
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                running_loss += loss.item() * inputs.size(0)
                preds = torch.argmax(outputs, dim=1)
                val_preds.extend(preds.cpu().numpy())
                val_labels.extend(labels.cpu().numpy())
        
        val_loss = running_loss / len(val_loader.dataset)
        val_acc = accuracy_score(val_labels, val_preds)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        # Сохранение лучшей модели
        if val_acc > best_acc:
            best_acc = val_acc
            best_model = model.state_dict().copy()
        
        print(f"\nEpoch {epoch+1} | Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f} "
              f"| Val Loss: {val_loss:.4f} Acc: {val_acc:.4f}")
    
    model.load_state_dict(best_model)
    return model, history

# Инициализация и обучение
set_random_state(random_state)
model = ImprovedNetwork().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
trained_model, history = train_model(model, train_loader, val_loader, criterion, optimizer, n_epochs=55)

# Сохранение модели
torch.save(trained_model.state_dict(), "/kaggle/working/best_model.pth")

# Визуализация
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train')
plt.plot(history['val_loss'], label='Validation')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history['train_acc'], label='Train')
plt.plot(history['val_acc'], label='Validation')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```
